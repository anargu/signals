```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
tweets = pd.read_csv("http://localhost:8080/tweets/s2020-06-01u2020-06-02.csv")
```

```python
sample_tweets = tweets.sample(int(1e3))
sample_tweets = sample_tweets.set_index(['created_at'])
sample_tweets.index = pd.to_datetime(sample_tweets.index, format='%Y-%m-%d %H:%M:%S')

sample_tweets = sample_tweets.groupby(pd.Grouper(freq='M')).first()
sample_tweets["created_at"] = sample_tweets.index

sns.relplot(x="created_at", y="replies_count", kind="line", aspect=3, data=sample_tweets)
```
